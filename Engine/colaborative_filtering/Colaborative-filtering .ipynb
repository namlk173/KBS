### Create funtion to evaluate accurate of matrix just make

```python
def get_similar(movieId, rating, corrMatrix_new, movies):
    try:
        similar_ratings = corrMatrix_new[movieId]*(rating-2.5)
        similar_ratings = similar_ratings.sort_values(ascending=False)
        return similar_ratings
    except:
        pass
```

```python
def get_list_rating(user_id):
    true_user_ratings = user_ratings_test.loc[user_id].to_numpy()
    full_ratings = []
    for i in range(0, len(true_user_ratings)):
        if true_user_ratings[i] != 0:
            full_ratings.append((columns_name[i], true_user_ratings[i]))
    split_ratings = []
    for i in range(0, int(0.1*len(full_ratings))):
        split_ratings.append(full_ratings[i])
    return split_ratings
```

```python
def predict(list_ratings):
    similar_movies = pd.DataFrame()
    for movie,rating in list_ratings:
        similar_movies = similar_movies.append(get_similar(movie,rating,corrMatrix_new, movies),ignore_index = True)
    return (similar_movies.sum().sort_values(ascending=False) + 2.5).to_numpy()
```

```python
def cal_accurate(true_user_ratings, predict):
    tong = 0
    dem = 0
    for i in range(0, len(true_user_ratings)):
        if true_user_ratings[i]!=0:
            tong = tong + 1
            if(predict[i] > 2.5 and true_user_ratings[i] >= 2.5 ) or (predict[i] <= 2.5 and true_user_ratings[i] < 2.5 ):
                dem = dem + 1
    return dem/tong
```

```python
def get_average_accurate():
    accurate = []
    for i in range(1, 21):
        true_user_ratings = user_ratings_test.loc[i].to_numpy()
        accurate.append(cal_accurate(true_user_ratings, predict(get_list_rating(i))))
    return accurate
```

```python
np.mean(get_average_accurate())
```

###  Create function to get recommend movies for uer

```python
def recommend(list_rating, corrMatrix_new, movies):
    similar_movies = pd.DataFrame()
    for movieId, rating in list_rating:
        similar_movies = similar_movies.append(get_similar(movieId, rating, corrMatrix_new, movies), ignore_index=True)
    list_index_movies = similar_movies.sum().sort_values(ascending=False).head(10).index
    list_movie_recommend = []
    for id in list_index_movies:
        list_movie_recommend.append(movies.loc[movies['movieId']==id])
    return list_movie_recommend
```

```python
list_recommend = recommend([(1,4), (2,4)], corrMatrix_new, movies)
```
